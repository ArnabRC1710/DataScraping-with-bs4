```python
# Let's obtain the links from the following website:
# https://www.whitehouse.gov/briefing-room/

# One of the things that this website consists of is the records of presidential briefings and statements

#Goal: Extract all of the links on that page that point to the briefings and statements
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
result = requests.get("https://www.whitehouse.gov/briefing-room/")
```

```python
print(result.status_code)
```

```python
print(result.headers)
```

```python
src = result.content
```

```python
src
```

```python
soup = BeautifulSoup(src,'lxml')
```

```python
print(soup.prettify)
```

```python
urls = []
for h2_tag in soup.find_all("h2"):
    a_tag = h2_tag.find('a')
    if a_tag != None:
        urls.append(a_tag.attrs['href'])
```
